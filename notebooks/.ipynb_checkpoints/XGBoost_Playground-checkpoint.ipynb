```python
#IMPORTS 
import kagglehub
import os
```

```python
from xgboost import XGBClassifier
# read data
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
data = load_iris()
X_train, X_test, y_train, y_test = train_test_split(data['data'], data['target'], test_size=.2)
# create model instance
bst = XGBClassifier(n_estimators=2, max_depth=2, learnbing_rate=1, objective='binary:logistic')
# fit model
bst.fit(X_train, y_train)
# make predictions
preds = bst.predict(X_test)
```

```python
# lets do a simple run on a SAT to GPA 

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer


path = kagglehub.dataset_download("luddarell/101-simple-linear-regressioncsv") # simple sat-gpa dataset 
csv_file_path = next((os.path.join(path, f) for f in os.listdir(path) if f.endswith('.csv')), None)
data = pd.read_csv(csv_file_path)

print(data.columns)

data.dropna(axis=0, subset=['GPA'], inplace=True) # drop nan rows 
y = data.GPA
X = data.drop(['GPA'], axis=1).select_dtypes(exclude=['object'])
train_X, test_X, train_y, test_y = train_test_split(X.values, y.values, test_size=0.25)

my_imputer = SimpleImputer()
train_X = my_imputer.fit_transform(train_X)
test_X = my_imputer.transform(test_X)
```

```python
from xgboost import XGBRegressor

my_model = XGBRegressor()
# Add silent=True to avoid printing out updates with each cycle
my_model.fit(train_X, train_y, verbose=False)
```

```python
predictions = my_model.predict(test_X)

from sklearn.metrics import mean_absolute_error
print("Mean Absolute Error : " + str(mean_absolute_error(predictions, test_y)))
```
