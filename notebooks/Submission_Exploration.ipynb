# In this we will look at the submission csv

```python
import pandas as pd

# Specify the file path to your CSV file
file_path = 'submission_format.csv'

# Read the CSV file into a DataFrame
sub_df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
print(len(sub_df))
display(sub_df.head(20))
```

```python
import pandas as pd
from datetime import datetime


# Extract airport code (first 4 characters before the underscore)
sub_df['airport_code'] = sub_df['ID'].str.extract(r'(^[A-Z]{4})')

# Extract date-time from the string
def parse_datetime(datetime_str):
    # Split by underscore and extract the date and time
    parts = datetime_str.split('_')
    date_str, time_str = parts[1], parts[2]  # Extract date and time

    # Convert to proper datetime format
    try:
        return datetime.strptime(date_str + time_str, '%y%m%d%H%M')
    except ValueError:
        # Return None if the datetime format doesn't match
        return None

# Apply the parsing function to create the datetime column
sub_df['datetime'] = sub_df['ID'].apply(parse_datetime)

# Display the resulting DataFrame
display(sub_df.head())
```

```python
import pandas as pd

sub_df['minutes_past'] = sub_df['ID'].str.split('_').str[-1].astype(int)
display(sub_df)
```

```python
# Define the bin edges and corresponding labels
bins = [0, 15, 30, 45, 60, 75, 90, 105, 120, 135, 150, 165, 180]  # Up to 180 minutes (3 hours)
labels = list(range(1, 13)) # Bin labels from 1 to 13

# Assign bins based on `minutes_past`, with 15 belonging to bin 2
sub_df['bin'] = pd.cut(sub_df['minutes_past'], bins=bins, labels=labels, right=True, include_lowest=True)

# Display the updated DataFrame
display(sub_df.head(20))
```
